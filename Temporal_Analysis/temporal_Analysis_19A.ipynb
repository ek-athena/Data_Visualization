```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame with years and numerical columns
df = pd.read_excel('/content/invasive_GPSC.xlsx')
df

# Set 'Year' as the index
df.set_index('Year', inplace=True)

# Calculate percentages
df_percentage = df.div(df.sum(axis=1), axis=0) * 100

# Plot the stacked bar plot
ax = df_percentage.plot(kind='bar', stacked=True, figsize=(10, 7))

# Adding labels and title
ax.set_ylabel('Percentage')
ax.set_xlabel('Year')
ax.set_title('Percentage of Invasive GPSCs')

# Display the legend
#plt.legend(title='Categories')
plt.legend(title='GPSC', bbox_to_anchor=(1, 1), loc='upper left')

# Show plot
plt.savefig('invasive_gpsc.svg')
plt.show()
```
